### XOR 게이트

```python
x_data=np.array([[0,0],[0,1],[1,0],[1,1]],dtype=np.float32)
y_data=np.array([[0],[1],[1],[0]],dtype=np.float32)
```

##### layer1

```python
X=tf.placeholder(tf.float32)
Y=tf.placeholder(tf.float32)
W1=tf.Variable(tf.random_normal([2,2]),name='weight1')  # [2,2] => [(x1,x2),출력값2개] : 즉 2개의 모형을 만들어냄
b1=tf.Variable(tf.random_normal([2]),name='bias1')      # [2]   => 출력값 2개
layer1=tf.sigmoid(tf.matmul(X,W1)+b1)
```

##### layer2

```python
W2=tf.Variable(tf.random_normal([2,1]),name='weight2')  # 2개의 모형을 입력받고 1개의 값을 출력
b2=tf.Variable(tf.random_normal([1]),name='bias2')      # 1개의 출력값
hypothesis=tf.sigmoid(tf.matmul(layer1,W2)+b2)          
```

### cost / loss function

```python
cost = -tf.reduce_mean(Y*tf.log(hypothesis)+(1-Y)*tf.log(1-hypothesis))
train = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(cost)
```

### Accuracy computation

#### True if hypothesis > 0.2 else False

```python
predicted = tf.cast(hypothesis > 0.2, dtype=tf.float32)     # 0.2 > 이면 1, 0.2 < 이면 0
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted,Y),dtype=tf.float32))
```

### Launch graph

```python
with tf.Session() as sess :
    sess.run(tf.global_variables_initializer())
    for step in range(10001) :
        sess.run(train, feed_dict={X:x_data,Y:y_data})
        if step % 500 == 0:     # 500번마다 프린트
            print("step :",step,"\ncost :",sess.run(cost,feed_dict={X:x_data,Y:y_data}),
                 "\nWeight :\n",sess.run([W1,W2]),'\nbias :\n',sess.run([b1,b2]))
            h, c, a = sess.run([hypothesis, predicted, accuracy], feed_dict={X:x_data,Y:y_data})
            print('Hypothesis:\n',h,'\nCorrect:\n',c,'\nAccuracy:',a)
```

### 1. 제공된 데이터(xor.txt)를 이용하여 XOR문제에 대한 실습코드를 작성하고 그 정확도를 측정하시오.

![image](https://user-images.githubusercontent.com/51112316/60855170-a9fac700-a23d-11e9-870d-a8c6a2e27d36.png)

```python
x_data=np.array([[0,0],[1,1],[1,0],[0,1]],dtype=np.float32)
y_data=np.array([[0],[0],[1],[1]],dtype=np.float32)
X=tf.placeholder(tf.float32)
Y=tf.placeholder(tf.float32)
```

#### 단일 포셉트론으로 비선형적 모델 해결 X

#### => layer 최소 2개 이상

```python
W1=tf.Variable(tf.random_uniform([2,2],-1.0,1.0))
W2=tf.Variable(tf.random_uniform([2,1],-1.0,1.0))

b1=tf.Variable(tf.zeros([2]),name='Bias1')
b2=tf.Variable(tf.zeros([1]),name='Bias2')
```

```python
L2=tf.sigmoid(tf.matmul(X,W1)+b1)
hypothesis=tf.sigmoid(tf.matmul(L2,W2)+b2)
```

```python
cost=-tf.reduce_mean(Y*tf.log(hypothesis)+(1-Y)*tf.log(1-hypothesis))
```

#### For this time, learning rate is very important!!!

```python
learning_rate=0.1
optimizer=tf.train.GradientDescentOptimizer(learning_rate)
train=optimizer.minimize(cost)
init=tf.global_variables_initializer()
```

```python
#Blas GEMM launch failed 오류시 아래 코드 사용
#gpu_options=tf.GPUOptions(per_process_gpu_memory_fraction=0.333)
#sess=tf.Session(config=tf.ConfigProto(gpu_options=gpu_options))
```

```python
with tf.Session() as sess:
    sess.run(init)
    
    for step in range(20001):
        sess.run(train, feed_dict={X:x_data,Y:y_data})
        if step % 500==0:
            print(step,sess.run(cost,feed_dict={X:x_data,Y:y_data}))
    
    #TEST MODEL
    correct=tf.equal(tf.floor(hypothesis+0.5),Y)
    #CALCULATE ACCURACY
    accuracy=tf.reduce_mean(tf.cast(correct,"float"))
    print(sess.run([hypothesis],feed_dict={X:x_data,Y:y_data}))
    print("Accuracy:",accuracy.eval({X:x_data,Y:y_data}))
```

#### 제공된 데이터를 이용하여 NAND 문제에 대한 실습 코드를 작성하고 그 정확도를 측정하시오.

```python
xy=np.loadtxt('NANDdata.txt',unpack=True)
```

```python
x_data=np.transpose(xy[0:-1])
y_data=np.reshape(xy[-1],(4,1))

X=tf.placeholder(tf.float32)
Y=tf.placeholder(tf.float32)
```

#### Define 2 layer Neural Network

```python
W1 = tf.Variable(tf.random_uniform([2,2],-1.0,1.0))
W2 = tf.Variable(tf.random_uniform([2,2],-1.0,1.0))

b1=tf.Variable(tf.zeros([2]),name='bias1')
b2=tf.Variable(tf.zeros([1]),name='bias2')
```

#### Our hypothesis

```python
L2=tf.sigmoid(tf.matmul(X,W1)+b1)
hypothesis=tf.sigmoid(tf.matmul(L2,W2)+b2)
```

#### Cross entropy cost function

```python
cost= -tf.reduce_mean(Y*tf.log(hypothesis)+(1-Y)*tf.log(1-hypothesis))
```

#### For this time, learning rate is very important.

```python
learning_rate=0.1
optimizer=tf.train.GradientDescentOptimizer(learning_rate)
train=optimizer.minimize(cost)
init=tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init)
    
    for step in range(10001):
        sess.run(train, feed_dict={X:x_data,Y:y_data})
        if step % 500==0:
            print(step,sess.run(cost,feed_dict={X:x_data,Y:y_data}))
    
    #TEST MODEL
    correct=tf.equal(tf.floor(hypothesis+0.5),Y)
    #CALCULATE ACCURACY
    accuracy=tf.reduce_mean(tf.cast(correct,"float"))
    print(sess.run([hypothesis],feed_dict={X:x_data,Y:y_data}))
    print("Accuracy:",accuracy.eval({X:x_data,Y:y_data}))
```
